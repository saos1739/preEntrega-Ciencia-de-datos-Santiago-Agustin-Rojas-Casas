### 4️⃣ Limpieza de datos

```python
# TODO: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados.
# Sugerencia: aplicar .drop_duplicates(), .str.strip(), .str.replace() según el caso.
# Documentar el proceso y los resultados.

#creamos las copias de los datasets originales

ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

#eliminamos las filas duplicadas

ventas_clean = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()

```

```python
#comprobamos que se haya realizado de forma correcta,

calidad(ventas_clean, 'Ventas(calidad)')
calidad(clientes_clean, 'Clientes(ingresos)')
calidad(marketing_clean, 'Marketing(producto)')

#hay 0 filas duplicadas, asi que se realizo de forma correcta
```

```python
# Normalizamos todo lo que sea texto con la funcion normalizar

def normalizar(df):
  for col in df.select_dtypes(include='object').columns:  # Filtramos el DF y nos quedamos con las columnas con Object
    df[col] = df[col].astype(str).str.strip().str.replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(" +", " ", regex=True).str.title()  #selecciona la columna actual y combierte el tipo de dato a string, borra los espacios en blanco o caracteres invisibles

  return df

  #normalizamos las fechas

for df in [ventas_clean, clientes_clean, marketing_clean]: #itera en cada df
  for col in df.columns: #itera en cada column
    if 'fecha' in col.lower(): #Busca las columnas con Fecha
      df[col]= pd.to_datetime(df[col], errors= 'coerce', dayfirst=True) #cambia el formato de las fechas a DD/MM/YY
```

```python
#aplicamos la funcion
ventas_clean = normalizar(ventas_clean)
clientes_clean = normalizar(clientes_clean)
marketing_clean = normalizar(marketing_clean)

display(ventas_clean)
display(clientes_clean)
display(marketing_clean)
```

```python
#Normalizamos los valores numerios

if 'precio' in ventas.columns:
  ventas_clean['precio'] = ventas_clean['precio'].astype(str).str.replace('$', '', regex=False).str.replace(',', '') # Primero convierte todo a texto, luego reemplaza los $ y los remplaza por espacios vacios, luego elimina todas las comas y quita los espacios sobrantes
  ventas_clean['precio'] = pd.to_numeric(ventas_clean['precio'], errors='coerce') # Convierte el texto a numero
```

```python
#nomalizamos Cantidad

if 'cantidad' in ventas_clean.columns:
  ventas_clean['cantidad'] = pd.to_numeric(ventas_clean['cantidad'], errors='coerce').astype('Int64') #Esto cambia el tipo de valor para que permita valores nulos

print(ventas_clean.dtypes)
```

```python
def filtro_rendimiento(dfs, nombres):
  resumen = []

  for df, nombres in zip(dfs, nombres):
    nulos = df.isna().sum().sum()
    duplicados = df.duplicated(keep= False).sum()
    col = len(df.columns)
    filas = len(df)
    resumen.append({"dataset": nombres,"filas": filas,"columna": col,"cant Nulos": nulos,"cant Duplicados": duplicados})

  report = pd.DataFrame(resumen)
  #print(report)
  return report

display(filtro_rendimiento([ventas_clean, clientes_clean, marketing_clean], ['ventas', 'clientes', 'marketing']))
```

### 5️⃣ Transformación de datos

```python
# TODO: Aplicar filtros y transformaciones para crear una tabla de ventas
# que muestre solo los productos con alto rendimiento. calcular el percentil 80
# y filtrar los productos que superen ese umbral en ventas.
#quantile(0.8)
# Sugerencia: usar .query() o condiciones con operadores lógicos.


def encontrar_columna(df, candidatos): #buscamos la columna
  for col in df.columns: #iteramos
    nombre = col.lower() #lower convierte todo a minuscula
    if any(candidato in nombre for candidato in candidatos): #si se cumple la condicion, nos devuelve el nombre de la columna
      return col
  return None #en caso contrario nos devuelve None
```

```python
#probamos la funcion

Product_col = encontrar_columna(ventas_clean, ['producto', 'precio'])
print(f'la columna es:', Product_col)
```

```python
# Calculamos el ingreso

ventas_perf = ventas_clean.assign(ingresos = ventas_clean['precio'] * ventas_clean['cantidad'])
```

```python
# Agregamos metricas

resumen_pro = ( ventas_perf.groupby(by= Product_col, dropna = False, as_index = False, observed = False).agg(ingreso_total=('ingresos', 'sum'),precio_promedio=('precio', 'mean'), registros=('ingresos', 'size')))
```

```python
print(resumen_pro.head(15))
```

```python
# Calculamos el percentil
p80 = resumen_pro['ingreso_total'].quantile(0.8, interpolation='linear')
print(f"{p80:,.2f}")
```

```python
#Filtramos los productos de alto rendimiento

ventas_top = resumen_pro.query(f'ingreso_total >= @p80', engine = "python").sort_values(['ingreso_total', 'registros'], ascending= [False, False], na_position='last', ignore_index= False )

print(f"Columna de producto detectada: {Product_col}")
print(f"percentil 80 de ingreso_total: {float(p80):,.2f}")
print("Productos de ALTO RENDIMIENTO:")
display(ventas_top.head(20))
```

### 6️⃣ Agregación

```python
# TODO: Resumir las ventas por categoría de producto y analizar los ingresos generados.
# Sugerencia: usar .groupby() y .agg() para generar métricas como suma y promedio.

def encontrar_columna(df, candidatos):
  for c in df.columns:
    if any(p in c.lower()  for p in candidatos):
      return c
  return None

#detectamos categoria
cat_col = encontrar_columna(ventas_clean, ['categoria'])
print(f'la columna es:', cat_col)

#Detectamos la columna ingreso

if 'ingresos' not in ventas_clean.columns:
  ventas_cat = ventas_clean.assign(ingresos = ventas_clean["precio"] * ventas_clean["cantidad"])
else:
    ventas_cat = ventas_clean.copy()


#Realizamos la agregacion

resumen_cat = (ventas_cat.groupby(by= cat_col, dropna= False, as_index= False)
               .agg(ingreso_total=('ingresos', 'sum'),
                    unidades=('cantidad', 'sum'),
                    ventas=('ingresos', 'size'),
                    precio_promedio=('precio', 'mean')))

resumen_cat = resumen_cat.sort_values(by='ingreso_total',ascending=False,na_position='last',ignore_index=True )

display(resumen_cat)

```
